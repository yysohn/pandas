## loc[]
- location을 의미하고
- 행인덱스, 컬럼의 이름으로 지정하는 방식
- loc[햄범위, '컬럼이름']
- loc[행범위, [['컬럼1', '컬럼2']]

```python
df.loc[:, '이름']
```

```python
df.loc[0:3, ['이름', '생년월일', '혈액형']]
```

```python
# df[행 Start 1:행 Stop 2] df[1:3]
# df.loc[행 Start 인덱스1 : 행 Stop 인덱스 3] df.loc[1:3]
df.loc[-1:2, ['이름', '생년월일']]
```

```python
# /행 : 2~5까지
# 컬럼 : 범위 지정은 '이름':'생년월일'
df.loc[2:5, ['이름', '생년월일']]
```

## 4. loc()와 isin() 활용
isin()을 활용한 색인은 정의한 list의 값이 있는 행만 추출하고자 할 때 사용

df.loc[df['컬럼명'].isin(['값1', '값2'])
df.loc[ df['컬럼명'].isin(['값1', '값2']), '추출 컬럼명' ]

```python
my_condition = ['플레디스', 'SM']
df.loc[ df['소속사'].isin(my_condition), ['소속사', '이름']]
```

```python
my_condition = ['방탄소년단']
df.loc[df['그룹'].isin(my_condition), ['이름', '소속사', '혈액형']]
```
