1. "datas/house_price.csv" 파일을 pandas DataFrame으로 loading 하기
2. 로딩한 DataFrame의 다음 정보 알아보기
- 컬럼명, 데이터갯수, 데이터 타입, 결측치 갯수를 알아보는 명령 수행하기
4. 로딩한 DataFrame 의 기술 통계 정보 확인하기
4. 컬럼명 바꾸기 : 분양가격(㎡) -> 분양가격
4. 두 개의 키(분양가, 연도)로 내림차순 소팅 하기
- 분양가가 가장 높았던 연도 확인

```python
data_file = "datas/house_price.csv"
df = pd.read_csv(data_file)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# 방법1 : df.columns : 전체 열 이름을 지정

df.columns = ['지역명', '규모구분', '연도', '월', '분양가격']
df
```

```python
# 방법2 : df.rename은 변경할 열 이름만 지정할 수 있음.

df.rename(columns={'분양가격': '분양가'}, inplace=True)
df
```

```python
# 문5) 분양가가 가장 높았던 연도 확인하기 위해 두 개의 키(분양가, 연도)로 내림차순 소팅하기
df.sort_values(by=['분양가', '연도'], ascending=False)
```
