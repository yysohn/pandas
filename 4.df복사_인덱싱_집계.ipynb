## 4. Groupby (그룹으로 묶어보기)

```python
import pandas as pd
df = pd.read_csv('datas/korea-star.csv')
df.head(3)
```

```python
df.value_counts('소속사')
```

```python
df.groupby('소속사')
```

```python
# 소속사별 각 컬럼의 갯수 카운트 값 보기(소속사 기준 groupby)
df.groupby('소속사').count()
```

```python
df.groupby('소속사').sum()
```

```python
# 밖의 [] : 컬럼명
# 안의 [,] : 리스트
df.groupby('소속사')[['키', '인기지수']].mean()
# df.groupby('소속사').mean(numeric_only=True)
```

```python
df.groupby('혈액형')[['키', '인기지수']].mean()
```

```python
# 그룹 by 수행수 multi index 데이터 프레임 만들고,
df2 = df.groupby(['혈액형', '성별'])[["키", "인기지수"]].mean()
df2
```

```python
df2 = df.groupby(['혈액형', '성별'])[["키", "인기지수"]].mean()
df2
```

```python
df2.stack()
```

```python
df2.unstack('성별')
```

```python
df2.unstack(['혈액형', '성별'])
```

```python
df2 = df2.reset_index()
df2
```
